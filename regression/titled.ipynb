### Cars

Take first 1000 records and save them to new file.

```python
cars = pd.read_csv("../datasets/vehicles.csv")
cars = cars.head(1000)[["price", "year", "model"]]
cars = cars.dropna()
cars[["price", "year", "model"]].to_csv("../datasets/vehicles-small.csv")
```

```python
cars = pd.read_csv("../datasets/vehicles-small.csv")[["price", "year", "model"]]
cars = cars[cars["year"].isin([2012.0, 2017.0, 2020.0])]
cars
```

```python
eins = np.ones((len(cars), 1))
X = np.column_stack((eins, cars["year"]))
y = cars["price"]
y
```

```python
# example
nd = [34, 40, 59, 85]
eins = np.ones((4, 1))
y = [1.3,2.9,3,6.5]
X = np.column_stack((eins, nd))
inverse = np.linalg.inv(np.matmul(np.transpose(X), X))
mult = np.matmul(inverse, np.transpose(X))
np.matmul(mult, y)
```

```python
inverse = np.linalg.inv(np.matmul(np.transpose(X), X))
mult = np.matmul(inverse, np.transpose(X))

teta = np.matmul(mult, np.transpose(y))
teta
```

```python
h = np.matmul(X, teta)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12, 8))

plt.plot(cars["year"], h, color="magenta")
plt.scatter(cars["year"], y)

plt.title("Linear regression")
plt.xlabel("Year: 2012, 2017, 2020", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.scatter(y, h)
z = np.polyfit(y, h, 1)
p = np.poly1d(z)
plt.plot(y, p(y), color="magenta")

plt.xlabel("Actual Labels", fontsize=12)
plt.ylabel("Predicted Label", fontsize=12)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

ols = LinearRegression().fit(X, y)
print(ols.coef_)
```

```python
predictions = ols.predict(X)
```

```python
plt.figure(figsize=(12, 8))

y = y.sort_values(axis=0)
predictions.sort()

plt.scatter(y, predictions)
z = np.polyfit(y, predictions, 1)
p = np.poly1d(z)
plt.plot(y, p(y), color="magenta")

plt.xlabel("Actual Labels", fontsize=12)
plt.ylabel("Predicted Labels", fontsize=12)
plt.show()
```

### Polynomial Regressions

```python
from sklearn.model_selection import train_test_split

y = cars["price"]

X_train, X_test, y_train, y_test = train_test_split(X[:,1], y, test_size=0.30, random_state=0)
ytest_sorted = y_test.sort_values(axis=0)
ytrain_sorted = y_train.sort_values(axis=0)
```

```python
import statsmodels.formula.api as smf
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly1 = PolynomialFeatures(degree=1, include_bias=False)
poly_features = poly1.fit_transform(X_train.reshape(-1, 1))
ols1 = LinearRegression().fit(poly_features, y_train)

predictions1_test = ols1.predict(poly1.transform(X_test.reshape(-1,1)))
predictions1_train = ols1.predict(poly_features)
```

```python
plt.figure(figsize=(12, 8))

plt.scatter(ytest_sorted, predictions1_test)
z = np.polyfit(ytest_sorted, predictions1_test, 1)
p = np.poly1d(z)
plt.plot(ytest_sorted, p(ytest_sorted), color="magenta")

plt.scatter(ytrain_sorted, predictions1_train)
z = np.polyfit(ytrain_sorted, predictions1_train, 1)
p = np.poly1d(z)
plt.plot(ytrain_sorted, p(ytrain_sorted), color="cyan")

plt.xlabel("Actual Labels", fontsize=12)
plt.ylabel("Predicted Labels", fontsize=12)
plt.show()
```
